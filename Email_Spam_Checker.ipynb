```python
import pandas as pd
```

```python
df = pd.read_csv("spam.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.groupby('Category').describe()
```

```python
df['spam'] = df['Category'].apply(lambda X:1 if X=='spam' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message,df.spam,test_size=0.3)
```

```python
len(X_test)
```

```python
!pip3 install scikit-learn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:2]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
emails = ['Hey Darling, can we get together to watch movie night tomorrow?','upto 20% discount on parking, exculsive offer just for you. Dont miss this reward!']
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count = v.transform(X_test)
```

```python
model.score(X_test_count,y_test)
```

Sklearn Pipeline

```python
from sklearn.pipeline import Pipeline
clf= Pipeline([('vectorizer',CountVectorizer()),('nb',MultinomialNB())])
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(emails)
```
